```python
import cv2    # OpenCV 
import matplotlib.pyplot as plt     # For displaying images in notebook
```

```python
# Read Image
img = cv2.imread('cat.jpeg')
```

```python
# image is stored in array (matrix) between 0 - 255
type(img)    
```

```python
# Height, width, color channels
img.shape
```

```python
# OpenCV uses "BGR".....not RGB
```

```python
# Correct way to display image
# BGR -> RGB

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
```

```python
# Save image to you computer
cv2.imwrite("saved_img.jpeg", img)
```
