```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("cat.jpeg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # BGR2GRAY removes 3 channels  convert to shape (height, width)

# cmap ="gray" : tells Matplotlib to - Map small value -> Black
#                                    - Map big values  -> White

plt.imshow(gray, cmap = 'gray')
plt.axis("off")
```

```python
# What if we not use cmap? 
plt.imshow(gray)
# it assumes default color map = "viridis" which assumes intensity 0 -> dark purple and 255 -> yello
```

## 1. Image Negavite

```python
# Negative Image
negative = 255 - gray
plt.imshow(negative, cmap ="gray")
```

```python
plt.figure(figsize =(10,4))
plt.subplot(1,2,1)
plt.imshow(gray, cmap="gray")
plt.title("Gray Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(negative, cmap="gray")
plt.title("Negative Image")
plt.axis('off')

plt.show()
```

## 2. Contrast Stretching

```python
img2 = cv2.imread("washed.jpg")
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
rmin = gray2.min()
rmax = gray2.max()


stretched = (gray2 - rmin)/(rmax - rmin) * 255
# Devision = converts into float =  image cannot store 123.15 intensity, so we convert back with uint8

# uint8 =  unsigned integer of 8 bits
# unsigned --> no negative values | Integer -- > whole numbers only | 8 bits --> Can represent 2^8 = 256 values [0 to 256] 
stretched = stretched.astype(np.uint8)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray2, cmap = 'gray')
plt.axis("off")
plt.title("Gray Image")

plt.subplot(1,2,2)
plt.imshow(stretched, cmap = 'gray')
plt.axis("off")
plt.title("Contrast Stretched Image")
plt.show()
```

## 3 Bit Plane Sclicing


```python
img3 = cv2.imread('cat.jpeg')
gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

plt.imshow(gray3, cmap = 'gray')
plt.axis('off')
```

```python
plt.figure(figsize = (12,4))

for i in range(8):
    plane = (gray3 >> i) & 1
    plt.subplot(2,4,i+1)
    plt.imshow(plane, cmap = 'gray')
    plt.title(f"Bit plane {i}")
    plt.axis('off')

plt.show()
```

## Gray Level Slicing

```python
# using for img3 and gray3

# Set the upper and lower 
lower = 100
upper = 150

# Make a sliced array with all values zero with same shape of gray3
sliced = np.zeros_like(gray3)

# sets value within range to 255(max)
sliced[(gray3>lower) & (gray3 < upper)] = 255

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray3, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sliced, cmap='gray')
plt.title("Gray Level Slicing")
plt.axis("off")

plt.show()
```
